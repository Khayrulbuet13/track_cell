```python
import matplotlib.pyplot as plt

def plot_deformation_index(file_path):
    # Read deformation index values from file
    with open(file_path, 'r') as file:
        deformation_indexes = [float(line.strip()) for line in file if line.strip()]

    # Generate a plot of the deformation indexes
    plt.figure(figsize=(5, 2.5), dpi=1200)  # Set the figure size
    plt.plot(deformation_indexes, marker='o', linestyle='-', color='b')  # Plot with markers and a line
    plt.title('Deformation Index Over Frames')  # Set the title of the plot
    plt.xlabel('Frame Number')  # Label the x-axis
    plt.ylabel('Deformation Index')  # Label the y-axis
    #plt.grid(True)  # Enable grid for better readability
    plt.show()  # Display the plot

# Specify the path to your DeformationIndex.txt file
file_path = 'T4_Notch_day1_B1_processed/Cell_157/DeformationIndex.txt'
plot_deformation_index(file_path)

```
